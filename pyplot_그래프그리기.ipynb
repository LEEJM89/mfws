#pyplot 으로 그래프 그리기

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,6,0.1)
# 0에서 6까지 , 0.1 간격으로 생성
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0,6,0.1)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.plot (x , y1 , label = "sin")
plt.plot (x , y2 , label = "cos")
```

```python
plt.xlabel("x")
plt.ylabel("y")
```

```python
plt.title('sin&cos')
```

```python
plt.plot(x,y1,label="sin")
plt.plot(x,y2,linestyle = "--", label="cos")

plt.legend()
plt.show()
```
